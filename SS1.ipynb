```python
import matplotlib.pyplot as plt

# Extract the data for plotting
longitude = data['longitude'].values
latitude = data['latitude'].values
climatology_last_month = data['climatology'][-1, :].values  # Last month's climatology

# Step 1: Adjust longitude values to be within -180 to 180 degrees
adjusted_longitude = ((longitude + 180) % 360) - 180

# Plotting again with adjusted longitudes
plt.figure(figsize=(10, 6))
plt.scatter(adjusted_longitude, latitude, c=climatology_last_month, cmap='viridis', s=1)
plt.colorbar(label='Climatology (°C)')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Climatology for the Last Month with Adjusted Longitudes')
plt.show()

```

```python
import xarray as xr



# Decodifica el tiempo si aún no está en un formato de fecha reconocible (esto puede no ser necesario si ya está en formato de fecha)
data['time'] = xr.cftime_range(start="1850", periods=data.dims['time'], freq='M', calendar='noleap')

# Calcular el promedio anual de la temperatura
# Esto agrupa los datos por año y luego calcula la media a lo largo de la dimensión de tiempo
temperatura_anual_promedio = data['temperature'].groupby('time.year').mean(dim='time')

# Cerrar el archivo NetCDF después de su uso
data.close()

# Ahora 'temperatura_anual_promedio' contiene la temperatura media para cada año y cada punto de la malla (latitud y longitud)

# Puedes convertir esto en un DataFrame de Pandas para un análisis más detallado o para almacenamiento/exportación
df_temperatura_anual_promedio = temperatura_anual_promedio.to_dataframe().reset_index()
```

```python
df_temperatura_anual_promedio
```

```python
import pandas as pd

# Asumiendo que df_temperatura_anual_promedio es tu DataFrame

# Agrupar por año y calcular la media de la temperatura, ignorando valores NaN
df_temperatura_media_por_año = df_temperatura_anual_promedio.groupby('year')['temperature'].mean().reset_index()

# Ahora df_temperatura_media_por_año tendrá solo dos columnas: 'year' y la temperatura media de ese año
df_temperatura_media_por_año
```

```python
import matplotlib.pyplot as plt

# Graficar la serie de tiempo del promedio anual de la temperatura
plt.figure(figsize=(20, 5))  # Configura el tamaño de la figura
plt.plot(df_temperatura_media_por_año['year'], df_temperatura_media_por_año['temperature'], marker='o')  # Crea un gráfico de líneas con marcadores

# Título y etiquetas
plt.title('Promedio Anual de Temperatura Global')
plt.xlabel('Año')
plt.ylabel('Temperatura Media (°C)')

# Mostrar una cuadrícula para facilitar la lectura
plt.grid(True)

# Opcional: Mostrar cada marca de año en el eje x
plt.xticks(df_temperatura_media_por_año['year'].values, rotation=45)  # Rota las etiquetas para que no se superpongan

# Mostrar el gráfico
plt.tight_layout()  # Ajusta automáticamente los parámetros de la subtrama
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Suponiendo que tienes un DataFrame llamado data con columnas 'lon', 'lat' para longitud y latitud
# data = pd.DataFrame({'lon': [-99.1332, -0.1257], 'lat': [19.4326, 51.5074], 'time': ['2021-01-01', '2021-01-02']})
# Define las regiones del AR6 utilizando regionmask
ar6_all = regionmask.defined_regions.ar6.all
# Crear una figura y un subplot con proyección
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_global()

# Añadir características al mapa
ax.add_feature(cfeature.LAND)
#ax.add_feature(cfeature.OCEANS)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Añadir los puntos de datos al mapa
scatter =ax.scatter(longitude, latitude,  c=climatology[-1], cmap='jet', s=10, transform=ccrs.Geodetic(), label='Ubicaciones')
cbar = plt.colorbar(scatter, ax=ax, orientation='horizontal', pad=0.05)
cbar.set_label('Temperatura (°C)')

ar6_all.plot(ax=ax, add_label=False)
# Añadir leyenda y título
ax.legend()
ax.set_title('Mapa de Ubicaciones')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
import cartopy.crs as ccrs
import cartopy.feature as cfeature


# Definir los límites de la malla sobre la que se interpolan los datos
grid_lon = np.linspace(data['longitude'].min(), data['longitude'].max(), 100)
grid_lat = np.linspace(data['latitude'].min(), data['latitude'].max(), 100)

# Crear malla de coordenadas
lon, lat = np.meshgrid(grid_lon, grid_lat)

# Interpolar los datos de temperatura sobre la malla
grid_temp = griddata((data['longitude'], data['latitude']), climatology[0], (lon, lat), method='cubic')

# Crear figura y eje con proyección de Cartopy
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([data['longitude'].min() - 5, data['longitude'].max() + 5, data['latitude'].min() - 5, data['latitude'].max() + 5], crs=ccrs.PlateCarree())

# Añadir características del mapa
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.COASTLINE)
ar6_all.plot(ax=ax, add_label=False)
# Visualizar la malla de temperatura
contour = ax.contourf(lon, lat, grid_temp, levels=100, cmap='RdYlBu_r', transform=ccrs.PlateCarree())

# Añadir barra de color
cbar = plt.colorbar(contour, ax=ax, orientation='horizontal', pad=0.05, aspect=50)
cbar.set_label('Temperatura (°C)')

# Añadir título
ax.set_title('Mapa de Temperatura')

plt.show()
```
