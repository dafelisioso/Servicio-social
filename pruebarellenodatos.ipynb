```python
import netCDF4 as nc
import pandas as pd
import numpy as np

# Rutas de tus archivos
netcdf_file = "Datos\Land_and_Ocean_EqualArea.nc"
tsv_file = "LandOcean_EqualArea_interpolated_at_10yr.tsv"
new_netcdf_file = "Datos\Land_and_Ocean_EqualArea_interpolado.nc"  # Nuevo archivo NetCDF

# Cargar archivo NetCDF original
dataset = nc.Dataset(netcdf_file, mode='r')

# Extraer las variables de interés
temperature = dataset.variables['temperature'][:]
latitudes = dataset.variables['latitude'][:]
longitudes = dataset.variables['longitude'][:]
time = dataset.variables['time'][:]

# Cargar archivo TSV
tsv_data = pd.read_csv(tsv_file, sep='\t')

# Asegurarse de que las dimensiones del archivo TSV coincidan con el NetCDF
if tsv_data.shape != temperature.shape:
    raise ValueError("Las dimensiones del archivo TSV no coinciden con las del archivo NetCDF")

# Reemplazar valores faltantes (por ejemplo, en NetCDF los valores faltantes suelen ser NaN o -9999)
missing_value = np.nan  # O el valor que indique los datos faltantes, por ejemplo -9999

# Convertir temperatura a un array de numpy para facilitar la manipulación
temperature_data = np.array(temperature)

# Reemplazar los valores faltantes con los del archivo TSV
for i in range(temperature_data.shape[0]):  # recorrer sobre el tiempo
    for j in range(temperature_data.shape[1]):  # recorrer sobre las latitudes
        for k in range(temperature_data.shape[2]):  # recorrer sobre las longitudes
            if np.isnan(temperature_data[i, j, k]):  # o "temperature_data[i, j, k] == missing_value"
                temperature_data[i, j, k] = tsv_data.iloc[j, k]  # asignar desde el TSV

# Crear un nuevo archivo NetCDF
new_dataset = nc.Dataset(new_netcdf_file, mode='w', format='NETCDF4')

# Crear las dimensiones en el nuevo archivo NetCDF
new_dataset.createDimension('time', len(time))
new_dataset.createDimension('latitude', len(latitudes))
new_dataset.createDimension('longitude', len(longitudes))

# Crear las variables en el nuevo archivo NetCDF
new_time = new_dataset.createVariable('time', 'f4', ('time',))
new_latitude = new_dataset.createVariable('latitude', 'f4', ('latitude',))
new_longitude = new_dataset.createVariable('longitude', 'f4', ('longitude',))
new_temperature = new_dataset.createVariable('temperature', 'f4', ('time', 'latitude', 'longitude'))

# Copiar los datos originales a las variables del nuevo archivo
new_time[:] = time
new_latitude[:] = latitudes
new_longitude[:] = longitudes
new_temperature[:] = temperature_data  # Temperatura con datos faltantes corregidos

# Cerrar el nuevo archivo NetCDF para guardar los cambios
new_dataset.close()
dataset.close()

print(f"Nuevo archivo NetCDF creado: {new_netcdf_file}")
```

```python
import xarray as xr
import pandas as pd
import numpy as np

# Load the NetCDF file
nc_file = "datos/Land_and_Ocean_EqualArea.nc"
ds = xr.open_dataset(nc_file)

# Load the TSV file, skipping the first row if it's a header
tsv_file = "LandOcean_EqualArea_interpolated_at_10yr.tsv"
tsv_data = pd.read_csv(tsv_file, sep="\t", header=0, usecols=range(2095))  # header=0 assumes first row is the header

# Ensure only numeric data is used, converting non-numeric to NaN and dropping such columns/rows
tsv_data = tsv_data.apply(pd.to_numeric, errors='coerce')

# Check the dimensions of the temperature variable (time, map_points)
temperature = ds['temperature']

# Find missing values in the temperature data
missing_values = np.isnan(temperature)

# Fill missing values from the corresponding TSV file
for t in range(temperature.shape[0]):  # Iterate over time dimension
    for p in range(temperature.shape[1]):  # Iterate over map_points
        if missing_values[t, p]:
            if not pd.isna(tsv_data.iloc[p, t]):  # Only fill if TSV data is valid
                # Replace missing temperature with the value from the TSV file
                temperature[t, p] = tsv_data.iloc[p, t]

# Save the updated NetCDF file
output_file = "datos/Updated_Land_and_Ocean_EqualArea.nc"
ds.to_netcdf(output_file)

print(f"Updated NetCDF file saved to {output_file}")
```
